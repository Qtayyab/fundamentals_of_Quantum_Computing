**X Gate**

---



Pauli x gate just flips the state of qubit or rotates it by 180 along x axis

```python
circuit = QuantumCircuit(1)
circuit.x(0)
circuit.draw(output = 'mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend = simulator).result()
statevector = result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector = }\n")
```

```python
plot_bloch_multivector(statevector)
```

As, then general state is prepared in zero state usually , but here output is 1 meaning it's invertig it.

Let's see another example

**Pauli Y Gate**


Pauli y gate also gives a shift in state along the y axis, but it is different from the x axis in thr manner that it also provide a shift in the phase of state. How,ever it is not possible to observe the difference physically on bloch sphere.

```python
circuit = QuantumCircuit(1)
circuit.y(0)
circuit.draw(output = 'mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend = simulator).result()
statevector = result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector = }\n")
```

As, it can be clearly seen that zero state of system has now inverted but not just inverted there is a iota there too due to phase change. but this effect is not obseraveable physically, such a bloch sphere below is of same as of x gate.

```python
plot_bloch_multivector(statevector)
```
